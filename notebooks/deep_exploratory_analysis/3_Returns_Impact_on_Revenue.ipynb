# RETURNS IMPACT ANALYSIS

> Reasoning: Highlight SKUs where high returns destroy margin. Focus on top_return_reason for potential pricing misalignment (if product quality perception doesn't match price).



```python
import plotly.express as px
import pandas as pd

data = pd.read_csv('finalweekly_modeldataset_cleaned.csv')
df = data.copy()
```

```python
'''Return Rate vs Margin'''
fig = px.scatter(df, x='return_rate', y='margin', color='is_high_return_sku',
                 hover_data=['top_return_reason'], title='Return Rate vs Margin')
fig.show()
```

## Insight
High return rate SKUs (>25%) are strongly associated with margin compression and volatility. Controlling returns in this range could preserve margins and stabilize profitability.

```python
'''Returns vs Revenue'''
fig = px.scatter(df, x='total_returns', y='weekly_revenue', color='category',
                 title='Returns vs Revenue')
fig.show()
```

# Insight
While most SKUs have low total returns (02), even a small number of returns can be associated with high-revenue SKUs. This points to a few high-return SKUs disproportionately contributing to revenue but potentially impacting long-term profitability and customer trust.
